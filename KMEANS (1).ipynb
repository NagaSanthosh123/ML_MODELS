```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
d=pd.read_csv('D:\ML_MODELS\Book1.csv')
d
```

```python
d.head()
```

```python
plt.scatter(d.Age,d['Income($)'])
plt.show()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
km
```

```python
y_pred=km.fit_predict(d[['Age','Income($)']])
y_pred
```

```python
d['cluster']=y_pred
d
```

```python
d1=d[d.cluster==0]
d2=d[d.cluster==1]
d3=d[d.cluster==2]

plt.scatter(d1.Age,d1['Income($)'],color='red')
plt.scatter(d2.Age,d2['Income($)'],color='green')
plt.scatter(d3.Age,d3['Income($)'],color='blue')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
l=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(d[['Age','Income($)']])
    l.append(km.inertia_)
```

```python
l
```

```python
l1=range(1,11)
plt.xlabel('K')
plt.ylabel('SUM OF SQUARED ERRORS')
plt.plot(l1,l)
```
